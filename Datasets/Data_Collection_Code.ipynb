# Data Collection

Collect at least 20 images of two categories: cats and dogs from the internet. I will be using the Bing Data Scraper for this.

```python
# Import necessary libraries
import os
from bing_image_downloader import downloader
```

I will collect 110 pictures for each class and use 100 images for training and 10 for testing. 

```python
# Create output directories
# Manually move 10 images from train directory to test directory

os.mkdir('train_set')
os.mkdir('test_set')
```

```python
# Check to see if the folders have been created
os.listdir()
```

```python
# Collect 110 dog images 
downloader.download(query='Dog face', 
                    limit=110,  
                    output_dir='train_set', 
                    adult_filter_off=True, 
                    force_replace=False)
```

```python
# Collect 110 cat images 
downloader.download(query='Cat', 
                    limit=110,  
                    output_dir='train_set', 
                    adult_filter_off=True, 
                    force_replace=False)
```

After this step, I will do some manual cleaning. If there is an image that I don't think will be great for training then I will remove it. I will keep equal number of dog and cat images.
